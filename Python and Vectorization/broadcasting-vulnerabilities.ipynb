Because broadcasting allows for shorthand operations that make vector math efficient and possible, incorrect implementations will not throw errors.

```python
import numpy as np

a = np.random.randn(5)
```

```python
print(a)
```

```python
print(a.shape)
```

(5,) <- this comma means it is called a "rank 1 array" which is neither a row vector or a column vector 

```python
print(a.T)
print(a.T.shape)
```

The shape is the same when you transpose.

```python
print(np.dot(a,a.T))
```

This returns a scalar instead of a matrix.  We should try to avoid rank 1 arrays by doing this

```python
a = np.random.randn(5,1)
print(a)
```

```python
print(a.T)
```

Notice the double square bracket ^^^ this means its a vector rather than this "rank 1 array"

```python
print(np.dot(a, a.T))
```

We should always declare the full shape of vectors.  One dimension should always be 1 in vectors.  If you do end up with a rank 1 array, you can redefine the shape as follows

```python
a = a.reshape(5,1)
```
