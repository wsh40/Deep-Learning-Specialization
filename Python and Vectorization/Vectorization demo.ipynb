```python
import numpy as np

a = np.array([1,2,3,4])
print(a)
```

```python
import time

#vectorized version
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print(c)
print("Vectorized version:" + str(1000*(toc-tic)) + "ms")

# For Loop version
c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()

print(c)
print("For loop:" + str(1000*(toc-tic))+ "ms")


# These two blocks do the exact same thing and return the same answer, though the vectorized version runs about 300x faster
# Instead of using a for loop, we're able to find the same answer by just doing a dot product
```
