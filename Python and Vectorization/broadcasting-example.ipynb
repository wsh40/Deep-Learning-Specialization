We create a vector which represents 4 different foods as columns (4) and the number of calories for each from carbs, fat, and protien as rows (3) 

```python
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0],
              [1.2, 104.0, 52.0, 8.0],
              [1.8, 135.0, 99.0, 0.9]])

print(A)
```

```python
cal = A.sum(axis=0)
print(cal)
```

```python
percentages = 100 * A / cal.reshape(1, 4)
print(percentages)
```
